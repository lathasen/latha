```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.model_selection import train_test_split
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
dataset = pd.read_csv("/content/mnist_test.csv")
dataset.head(10)
```

```python
dataset.describe()
```

```python
dataset.dtypes
```

```python
X = dataset.iloc[:,:20].values
y = dataset.iloc[:,20:21].values
```

```python
X
```

```python
y
```

```python
# handling the features for NN model
sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
# handling the lable
ohe = OneHotEncoder()
y = ohe.fit_transform(y).toarray()
print(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size= 0.1, random_state=12)
```

```python
model = Sequential()
model.add(Dense(units = 16,
activation='relu',
use_bias=True,
kernel_initializer="glorot_uniform",
bias_initializer="zeros",
kernel_regularizer=None,
bias_regularizer=None,
activity_regularizer=None,
kernel_constraint=None,
bias_constraint=None,
input_dim = 20))
model.add(Dense(units = 12,
activation='relu',
use_bias=True,
kernel_initializer="glorot_uniform",
bias_initializer="zeros",
kernel_regularizer=None,
bias_regularizer=None,
activity_regularizer=None,
kernel_constraint=None,
bias_constraint=None,
))
model.add(Dense(units = 4,
activation='softmax',
use_bias=True,
kernel_initializer="glorot_uniform",
bias_initializer="zeros",
kernel_regularizer=None,
bias_regularizer=None,
activity_regularizer=None,
kernel_constraint=None,
bias_constraint=None,
))
```

```python
type(model)
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'],loss_weights=None,
              weighted_metrics=None,run_eagerly=None,steps_per_execution=None,
              jit_compile=None)
```

```python
type(model)
```

```python
y_pred = model.predict(X_test)
```

```python
pred = list()
for var in range(len(y_pred)):
  pred.append(np.argmax(y_pred[var]))
```

```python
test = list()
for var in range(len(y_test)):
  test.append(np.argmax(y_test[var]))
```

```python
from sklearn.metrics import accuracy_score
score = accuracy_score(pred,test)
print("Accurcy of the ANN Model is : ", score*100, "%")
```
